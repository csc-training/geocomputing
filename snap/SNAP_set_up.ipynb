As example see what plugins are available, can be left out from user scripts.

```python
ProductIOPlugInManager = jpy.get_type(
    "org.esa.snap.core.dataio.ProductIOPlugInManager"
)

manager = ProductIOPlugInManager.getInstance()
readers = manager.getAllReaderPlugIns()

while readers.hasNext():
    r = readers.next()
    print(r)
```
